```python
df = pd.read_csv("C:/Users/vadim/Downloads/titanic.csv")
```

```python
df_female = df[df['Sex'] == 'female']
df_male = df[df['Sex'] == 'male']
```

```python
df_cleaned = df.dropna()
df_male = df_male.dropna()
df_female = df_female.dropna()
```

```python
import matplotlib.pyplot as plt

df['Age'].plot(kind='hist', bins=100, alpha=0.7, color="pink")
plt.xlabel('Возраст')
plt.ylabel('Количество пассажиров')
plt.grid(axis='y')
```

```python
def calc(df):
    return {
        'min': df['Age'].min(),
        'max': df['Age'].max(),
        'mean': df['Age'].mean(),
        'mode': df['Age'].mode()[0],
        'median': df['Age'].median()
    }

stats_male = calc(df_male)
stats_female = calc(df_female)

pd.DataFrame({
    'Male': stats_male,
    'Female': stats_female
})
```

```python
df_male['Age'].plot(kind='line', label='Male', color='blue', linestyle='-')
df_female['Age'].plot(kind='line', label='Female', color='red', linestyle='-')

plt.title('Линейный график')
plt.ylabel('Возраст')
plt.legend()
plt.show()
```
